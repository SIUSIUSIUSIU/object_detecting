```python
# 10,000 프레임 이하의 영상만 사용한다고 가정했을 때
def make_name(num):
    if num < 10: return '000'+str(num)
    elif num < 100: return '00'+str(num)
    elif num < 1000: return '0'+str(num)
    else: return str(num)
```

```python
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
import numpy as np

file_name = 'Your FileName'

# Making Video
cap = cv2.VideoCapture(file_name)
video_output_path = '/content/drive/MyDrive/detect_result.mp4'
# 저장할 비디오 파일의 코덱 
codec = cv2.VideoWriter_fourcc(*'XVID')

#비디오 화면 크기
#cv2.CAP_PROP_FRAME_WIDTH : file_name 비디오 프레임의 가로
#cv2.CAP_PROP_FRAME_HEIGHT) : file_name 비디오 프레임의 세로
vid_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

#file_name 비디오의 초당 프레임
vid_fps = cap.get(cv2.CAP_PROP_FPS)

#cv2.VideoWriter : 비디오 파일을 생성해서 저장할 객체 생성
#cv2.VideoWriter(저장할 경로, 비디오코덱, 비디오 초당 프레임, 프레임크기)    
vid_writer = cv2.VideoWriter(video_output_path, codec, vid_fps, vid_size) 
```

```python
cap = cv2.VideoCapture(file_name)
frame_num = 1
while True:
    ret, frame = cap.read()
    if frame is None:
        break

    #   # 1초에 한번씩 바운딩 박스를 칠 때 사용하는 코드
    # if frame_num % vid_fps == 0:
    #     img = Image.fromarray(np.array(frame), 'RGB')
    #     img.save('/content/yolov7/custom_test/frame_'+make_name(frame_num)+'.jpg')

    # 모든 프레임에 바운딩 박스를 칠 때 사용하는 코드
    img = Image.fromarray(np.array(frame[:,:,::-1]))
    img.save('/content/yolov7/custom_test/frame_'+make_name(frame_num)+'.jpg')

    frame_num += 1
    
cap.release()
```
